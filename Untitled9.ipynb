```python
def checkValidParenthesis(string):
    global validParenthesis
    stack = []
    
    for i in string:
        if i == '(':
            stack.append(i)
        elif i == ')':
            if len(stack) == 0:
                stack.append(i)
                return
            elif stack[-1] == '(':
                stack.pop()
    validParenthesis.append(string)
```

```python
import itertools

def foo(l):
    yield from itertools.permutations(l,n*2)

result = []
for x in foo('()'*n):
    result.append(''.join(x))
```

```python
validParenthesis = []
for i in result:
    checkValidParenthesis(i)
```
