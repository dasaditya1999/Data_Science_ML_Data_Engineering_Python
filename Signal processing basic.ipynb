```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from math import pi
```

```python
time = np.arange(-1,5,000.1)
k = 4/pi
m = pi/2
```

```python
m*time
```

```python
k1 = k*np.cos(m*time)
k2 = (-k/3)*np.cos(3*m*time)
k3 = (k/5)*np.cos(5*m*time)
k4 = (-k/7)*np.cos(7*m*time)
k5 = (k/9)*np.cos(9*m*time)

amplitude = k1 + k2 + k3 + k4 + k5
```

```python
amplitude
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1)

ax.plot(time, amplitude)
ax.set(xlabel='Time', ylabel='Amplitude')
```
